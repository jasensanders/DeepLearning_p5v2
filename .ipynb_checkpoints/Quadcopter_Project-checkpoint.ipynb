## Controlling the Quadcopter

We provide a sample agent in the code cell below to show you how to use the sim to control the quadcopter.  This agent is even simpler than the sample agent that you'll examine (in `agents/policy_search.py`) later in this notebook!

The agent controls the quadcopter by setting the revolutions per second on each of its four rotors.  The provided agent in the `Basic_Agent` class below always selects a random action for each of the four rotors.  These four speeds are returned by the `act` method as a list of four floating-point numbers.  

For this project, the agent that you will implement in `agents/agent.py` will have a far more intelligent method for selecting actions!

```python
import random

class Basic_Agent():
    def __init__(self, task):
        self.task = task
    
    def act(self):
        new_thrust = random.gauss(450., 25.)
        return [new_thrust + random.gauss(0., 1.) for x in range(4)]
```

Run the code cell below to have the agent select actions to control the quadcopter.  

Feel free to change the provided values of `runtime`, `init_pose`, `init_velocities`, and `init_angle_velocities` below to change the starting conditions of the quadcopter.

The `labels` list below annotates statistics that are saved while running the simulation.  All of this information is saved in a text file `data.txt` and stored in the dictionary `results`.  

```python
%load_ext autoreload
%autoreload 2

import csv
import numpy as np
from task import Task

# Modify the values below to give the quadcopter a different starting position.
runtime = 5.                                     # time limit of the episode
init_pose = np.array([0., 0., 10., 0., 0., 0.])  # initial pose
init_velocities = np.array([0., 0., 0.])         # initial velocities
init_angle_velocities = np.array([0., 0., 0.])   # initial angle velocities
file_output = 'data.txt'                         # file name for saved results

# Setup
task = Task(init_pose, init_velocities, init_angle_velocities, runtime)
agent = Basic_Agent(task)
done = False
labels = ['time', 'x', 'y', 'z', 'phi', 'theta', 'psi', 'x_velocity',
          'y_velocity', 'z_velocity', 'phi_velocity', 'theta_velocity',
          'psi_velocity', 'rotor_speed1', 'rotor_speed2', 'rotor_speed3', 'rotor_speed4']
results = {x : [] for x in labels}

# Run the simulation, and save the results.
with open(file_output, 'w') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(labels)
    while True:
        rotor_speeds = agent.act()
        _, _, done = task.step(rotor_speeds)
        to_write = [task.sim.time] + list(task.sim.pose) + list(task.sim.v) + list(task.sim.angular_v) + list(rotor_speeds)
        for ii in range(len(labels)):
            results[labels[ii]].append(to_write[ii])
        writer.writerow(to_write)
        if done:
            break
```

## The Task

A sample task has been provided for you in `task.py`.  Open this file in a new window now. 

The `__init__()` method is used to initialize several variables that are needed to specify the task.  
- The simulator is initialized as an instance of the `PhysicsSim` class (from `physics_sim.py`).  
- Inspired by the methodology in the original DDPG paper, we make use of action repeats.  For each timestep of the agent, we step the simulation `action_repeats` timesteps.  If you are not familiar with action repeats, please read the **Results** section in [the DDPG paper](https://arxiv.org/abs/1509.02971).
- We set the number of elements in the state vector.  For the sample task, we only work with the 6-dimensional pose information.  To set the size of the state (`state_size`), we must take action repeats into account.  
- The environment will always have a 4-dimensional action space, with one entry for each rotor (`action_size=4`). You can set the minimum (`action_low`) and maximum (`action_high`) values of each entry here.
- The sample task in this provided file is for the agent to reach a target position.  We specify that target position as a variable.

The `reset()` method resets the simulator.  The agent should call this method every time the episode ends.  You can see an example of this in the code cell below.

The `step()` method is perhaps the most important.  It accepts the agent's choice of action `rotor_speeds`, which is used to prepare the next state to pass on to the agent.  Then, the reward is computed from `get_reward()`.  The episode is considered done if the time limit has been exceeded, or the quadcopter has travelled outside of the bounds of the simulation.

In the next section, you will learn how to test the performance of an agent on this task.

## The Agent

The sample agent given in `agents/policy_search.py` uses a very simplistic linear policy to directly compute the action vector as a dot product of the state vector and a matrix of weights. Then, it randomly perturbs the parameters by adding some Gaussian noise, to produce a different policy. Based on the average reward obtained in each episode (`score`), it keeps track of the best set of parameters found so far, how the score is changing, and accordingly tweaks a scaling factor to widen or tighten the noise.

Run the code cell below to see how the agent performs on the sample task.

```python
import sys
import pandas as pd
from agents.policy_search import PolicySearch_Agent
from task import Task

num_episodes = 1000
target_pos = np.array([0., 0., 10.])
task = Task(target_pos=target_pos)
agent = PolicySearch_Agent(task) 

for i_episode in range(1, num_episodes+1):
    state = agent.reset_episode() # start a new episode
    while True:
        action = agent.act(state) 
        next_state, reward, done = task.step(action)
        agent.step(reward, done)
        state = next_state
        if done:
            print("\rEpisode = {:4d}, score = {:7.3f} (best = {:7.3f}), noise_scale = {}".format(
                i_episode, agent.score, agent.best_score, agent.noise_scale), end="")  # [debug]
            break
    sys.stdout.flush()
```

```python
## TODO: Train your agent here.
import sys
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from agents.agent import DDPG
from task import TakeOffTask

num_episodes = 1500
target_pos = np.array([0., 0., 20.])
init_pos = np.array([0., 0.,10., 0.,0.,0.])
task = TakeOffTask(init_pose=init_pos, target_pos=target_pos)
agent = DDPG(task) 
labels = ['time', 'x', 'y', 'z', 'phi', 'theta', 'psi', 'x_velocity',
          'y_velocity', 'z_velocity', 'phi_velocity', 'theta_velocity',
          'psi_velocity', 'rotor_speed1', 'rotor_speed2', 'rotor_speed3', 'rotor_speed4', 'total_reward','reward']

fileLabel = "data_Episode_"
fileExt = ".csv"
directory = "Episode_Data/"
# Convetion v = Network Version, r = Reward Version, L = longer Run 
networkVersion = "v3r6"


plot_results_every = 50 # episodes.  

for i_episode in range(1, num_episodes+1):
    
    # Setup CSV file to write episode data
    file_output = directory + fileLabel + "{:04d}".format(i_episode) + networkVersion + fileExt
    #file_output = directory + "data.csv"
    
    
    state = agent.reset_episode() # start a new episode
    
    # Storage for row data
    rows = []
    
    while True:
        action = agent.act(state) 
        next_state, reward, done = task.step(action)
        agent.step(action, reward, next_state, done)
        state = next_state
            
        # Prep for write to file
        to_write = [task.sim.time] + list(task.sim.pose) + list(task.sim.v) + list(task.sim.angular_v) + list(action) + [agent.total_reward] + [reward]
        rows.append(to_write)
            
        if done:
            print("\rEpisode = {:4d}, Average Reward = {:7.3f} (Total Reward = {:7.3f}), X = {:7.3f}, Y = {:7.3f}, Z = {:7.3f}, actions = {:4d}".format(
                i_episode, agent.avg_reward, agent.total_reward, task.sim.pose[0], task.sim.pose[1], task.sim.pose[2], agent.count), end="")
            break
    # Write a log every so many episodes
    if i_episode%plot_results_every ==0:
        with open(file_output, 'w') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(labels)
            for i in range(len(rows)):
                writer.writerow(rows[i])
                    
    sys.stdout.flush()
```

## Plot the Rewards

Once you are satisfied with your performance, plot the episode rewards, either from a single run, or averaged over multiple runs. 

```python
## TODO: Plot the rewards.
import csv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

labels = ['time', 'x', 'y', 'z', 'phi', 'theta', 'psi', 'x_velocity',
          'y_velocity', 'z_velocity', 'phi_velocity', 'theta_velocity',
          'psi_velocity', 'rotor_speed1', 'rotor_speed2', 'rotor_speed3', 'rotor_speed4', 'total_reward','reward']
fileLabel = "data_Episode_"
fileExt = ".csv"
directory = "Episode_Data/"
networkVersion = "v2r6L"
file_output = directory + fileLabel + "0500" + networkVersion + fileExt
plot_results_every = 50

def plotEpisode(epNum, netVer):
    epNumber = "{:04d}".format(epNum)
    file_output = directory + fileLabel + epNumber + netVer + fileExt
    
    with open(file_output, mode='r') as infile:

        resultsC = {x : [] for x in labels}
        reader =csv.reader(infile)
        count = 0
        for row in reader:
        
            if len(row) != 0 and count !=0:
                for i,label in enumerate(labels):
                    resultsC[label].append(float(row[i]))
                    
            count+=1
    return resultsC

resultsC = plotEpisode(1500, "v3r6")

def plotEpisodes(start, end, netVer):
    labelsRewards = ['episode', 'x', 'y', 'z', 'phi', 'theta', 'psi', 'x_velocity',
          'y_velocity', 'z_velocity', 'phi_velocity', 'theta_velocity',
          'psi_velocity', 'rotor_speed1', 'rotor_speed2', 'rotor_speed3', 'rotor_speed4', 'total_reward','reward']
    resultsE = {x: [] for x in labelsRewards}
    for i in range(start, end+plot_results_every, plot_results_every):
        # We are taking the last row of every episode in the range to map the rewards, values.
        episodei = "{:04d}".format(i)
        file_output = directory + fileLabel + episodei + netVer + fileExt
        with open(file_output, mode='r') as infile:
            rows=[]
            reader =csv.reader(infile)
            count = 0
            # grab all the rows in the episode
            for row in reader:
                if len(row) != 0 and count !=0:
                    rows.append(row)
                count += 1
            # Get the last row
            lastRow = rows[-1]
            #print("Last Row Length: ",len(lastRow), "\n")
            # Add to dictionary of episode data
            for j, label in enumerate(labelsRewards):
                if(j ==0):
                    resultsE[label].append(float(i))
                else:
                    resultsE[label].append(float(lastRow[j]))
        
    return resultsE
            
            
resultsE = plotEpisodes(50,1500, "v3r6")
   

    
```

**Question 1**: Describe the task that you specified in `task.py`.  How did you design the reward function?

**Answer**:  **The Task :** I started with the default task, and modified it to be a "take off" task starting at z=10 targeting z=20. After some initial tests with the defaults and reading the slack channel, it was suggested that I not start at z=0 because it would not give the agent very much time to learn. It would end the episode before the rewards function and learning could take effect.
This proved true as my number of actions went up after making the change.  I added targets for velocity, Angular velocity, and angles. 

**The Reward Function :**
The rewards function proved to be the single hardest thing to master.  Knowledge of the physics of the problem you are trying to solve is super important.  For example if you are given a euler angle in radians it is important to know why you might want the sine and cosine of it and why that might be useful in your rewards funtion. The slack channel mentioned triginomic identities, so Khan Academy to the rescue.   In my code you can see the evolution of my rewards function.  

In version 1 I rewarded z or upward velocity and penalized off angles, velocities and positions. I then decided to normalize the out put of the reward to between -1 and 1 using tanh. That actually helped alot, but the performace was still to erratic.

In version 1.5 I figured out that the cosine of the angles were important, so i added that. 

In version 2 I upped the penalty on angular velocities.  That killed perfomance. Dead.  

In version 3 I got rid of the cosine and switched to a different reward function. 1/x^2.  I liked it because as the differences got closer to zero, the rewards went up.  Although the implementation in the code (1/0.02+x^2) didn't work, I liked where it was going. A math relationship where the input could be positive or negative and as the input got closer to zero the reward goes up. 12/x^2 would have been better after taking the sine of the radians to get the input. I may try that in the future. 

In version 4 (After getting through Triginomic Identities on Khan Academy) I realized that the sine of the angles was important for judging error as well so i went back to using the cosine and added the sine as well. It was at this point that I dicovered the DDPG torcs project and sure enough, the same type of reward function using sine and cosine of the radians was there too, so i knew I was on the right track.  I also found a great description  of OU noise that helped my understanding of why you might want to change those perameters and what they do. I also stopped taking the tanh of the reward as it just seemed to make verything take longer. Performace improved, but still had fits and starts.  The Theta angle was my problem.

In version 5 I decided to single out Theta and penalize it on top of the summed angles penalty. This just threw everything off. It did not go well.

In version 6 I elimnated the Theta penalty and just added a penalty for off position. I also added a severe penalty for crashing. Because this is a "take off" task I am not really concerned about it maintaining the target height more than a consistant level flight upwards.  Maintaining would be to hover.  I'll try that later.

This seemed to make sense for this case because the act of balancing on air is very sensitive so large changes may make you overcompensate.  The Default Learing rate for the Adam optimizer is 0.001 which I changed to 0.0001 for the actor.(Critic remained at default 0.001)

The gamma rate I left at 0.99

The Tau though I modified.  When I was training it seemed as if i was getting somewhere and then the rewards would fall off a cliff and it would crash about 200 episodes in to training. I want the quadcopter to be able to make actions to correct, just not so big that they endanger stability.  This seemed to help.

For I left the network structure largely as is, except for numbers of hidden units. I doubled those both in the actor and the critic to 64, 128 from 32,64.
